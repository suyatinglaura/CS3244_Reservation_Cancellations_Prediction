```python
import pandas as np
import matplotlib.pyplot as plt
import pandas as pd

# Creating a mock DataFrame
data = {
    'Model': ['Logistic Regression', 'Decision Tree'],
    'Accuracy_Pre_Tuning': [0.78, 0.71],
    'Accuracy_Post_Tuning': [0.85, 0.75]
}
metrics_df = pd.DataFrame(data)

# Extracting data for plotting
models = metrics_df['Model']
accuracy_pre_tuning = metrics_df['Accuracy_Pre_Tuning']
accuracy_post_tuning = metrics_df['Accuracy_Post_Tuning']

x = np.arange(len(models))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 5))
rects1 = ax.bar(x - width/2, accuracy_pre_tuning, width, label='Pre-Tuning')
rects2 = ax.bar(x + width/2, accuracy_post_tuning, width, label='Post-Tuning')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Model')
ax.set_title('Model Accuracy Before and After Hyperparameter Tuning')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create mock data
data = {'Model': ['Decision Tree', 'SVM', 'Adaboost', 'Gradient Boost', 'LightGBM', 'Random Forest', 'KNN', 'Logistic Regression'],
        'Accuracy_Before': [0.45577130528586834, 0.46669175762137755, 0.485190, 0.5185972369819342, 0.48156342182890854, 0.433707, 0.4989165763813651, 0.5010879701585328],
        'Accuracy_After': [0.515308, 0.49915014164305943, 0.525457, 0.5503355704697986, 0.5533419023136246, 0.538363, 0.4923887587822014, 0.508391821788221]}

df = pd.DataFrame(data)

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Set the bar width and positions
bar_width = 0.35
positions = np.arange(len(df))

# Create the bars
bars1 = ax.bar(positions - bar_width/2, df['Accuracy_Before'], bar_width, label='Before Tuning')
bars2 = ax.bar(positions + bar_width/2, df['Accuracy_After'], bar_width, label='After Tuning')

# Set the x-axis tick labels
ax.set_xticks(positions)
ax.set_xticklabels(df['Model'])

# Add labels and title
ax.set_xlabel('Model')
ax.set_ylabel('F1 Sc0re')
ax.set_title('Comparison of Model Accuracy Before and After Hyperparameter Tuning')
all_accuracies = np.concatenate([df['Accuracy_Before'], df['Accuracy_After']])
y_min = all_accuracies.min() * 0.98
y_max = all_accuracies.max() * 1.02
ax.set_ylim([y_min, y_max])# Add legend
ax.legend()

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
